```python
import numpy as np

# Define the path to the GloVe embeddings file
embedding_file_path = "/content/glove.6B.50d.txt"

# Define the words for which you want to calculate similarity
word1 = "king"
word2 = "queen"

# Function to load GloVe embeddings
def load_glove_embeddings(embedding_file):
    embeddings = {}
    with open(embedding_file, "r", encoding="utf-8") as file:
        for line in file:
            values = line.split()
            word = values[0]
            embedding = np.array(values[1:], dtype=float)
            embeddings[word] = embedding
    return embeddings

# Load the GloVe embeddings
glove_embeddings = load_glove_embeddings(embedding_file_path)

# Check if the words exist in the GloVe embeddings
if word1 in glove_embeddings and word2 in glove_embeddings:
    # Calculate the Euclidean distance between the two words' embeddings
    embedding1 = glove_embeddings[word1]
    embedding2 = glove_embeddings[word2]
    euclidean_distance = np.linalg.norm(embedding1 - embedding2)

    # Convert Euclidean distance to a similarity score
    similarity_score = 1 / (1 + euclidean_distance)

    print(f"Euclidean Distance between '{word1}' and '{word2}': {euclidean_distance:.4f}")
    print(f"Similarity Score between '{word1}' and '{word2}': {similarity_score:.4f}")

    # Print the vector representations of the words
    print(f"Vector representation of '{word1}':\n{embedding1}")
    print(f"Vector representation of '{word2}':\n{embedding2}")
else:
    print("One or both of the words are not found in the GloVe embeddings.")
```

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to normalize a vector
def normalize_vector(vector):
    norm = np.linalg.norm(vector)
    if norm == 0:
        return vector
    return vector / norm

# Normalize the embeddings
# normalized_embeddings = {word: normalize_vector(embedding) for word, embedding in glove_embeddings.items()}
# Extract the first 100 items from glove_embeddings
first_100_glove_embeddings = {word: embedding for i, (word, embedding) in enumerate(glove_embeddings.items()) if i < 50}

# Normalize the embeddings
normalized_embeddings = {word: normalize_vector(embedding) for word, embedding in first_100_glove_embeddings.items()}

# Extract the normalized embeddings as a matrix
embedding_matrix = np.array(list(normalized_embeddings.values()))

# Perform PCA to reduce dimensionality to 3D
pca = PCA(n_components=3)
embeddings_3d = pca.fit_transform(embedding_matrix)

# Create a 3D scatter plot with a larger figure size
fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111, projection='3d')

# Increase the marker size for scatter plot points
scatter = ax.scatter(embeddings_3d[:, 0], embeddings_3d[:, 1], embeddings_3d[:, 2], s=100)

# Add labels for each point
for i, (word, embedding) in enumerate(normalized_embeddings.items()):
    x, y, z = embeddings_3d[i]
    ax.text(x, y, z, word, fontsize=8)

# Set labels and title
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('3D Word Embedding Visualization')

# Show the plot
plt.show()
```
